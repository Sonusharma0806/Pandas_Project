###  Representation Of Total Case _Registered W.R.T States

```python
plt.figure(figsize=(12,8))
sns.barplot(data=df, x="No of Police Personel_Convicted", y="State_UT")
plt.xlabel("States")
plt.ylabel("Number of cases Police Personel_Convicted")
plt.title("Number of cases Police Personel_Convicted w.r.t States")
plt.grid(True)
```

### Total Number of Chargesheeted, Final Report And Police Personel_Convicted Cases

```python
case = df[["Cases_Chargesheeted","Cases_Final Report Submitted","No of Police Personel_Convicted"]].sum()
case
```

### Pie Chart Reprentation  Of  An Action Taken On the Basis Of Registered_Case(Over All)

```python
plt.figure(figsize=(7,7))
case.plot.pie(autopct="%1.2f%%",explode=(0.1,0,0))
plt.title("Action Taken On the Basis Of Registered_Case(Over All)")

plt.show()
```

### Adding Column for % of Chargesheeted Cases ( each State )

```python
df["Case_Chargesheeted_%"]=(df["Cases_Chargesheeted"]/df["Case _Registered"])*100
df.head(2)
```

### Adding Column for % of Police Personel_Convicted (each State )

```python
df["Case_Convicted_%"]=(df["No of Police Personel_Convicted"]/df["No of Police_Chargesheeted"])*100
df.head()
```

### Viewing State with  Maximum % Of Chargesheeted Cases 

```python
df[df['Case_Chargesheeted_%']==(df["Case_Chargesheeted_%"].max())][['State_UT',"Case _Registered","Case_Chargesheeted_%"]]
```

### Viewing State with  Minimum % Of Chargesheeted Cases 

```python
df[df['Case_Chargesheeted_%']==(df["Case_Chargesheeted_%"].min())][['State_UT',"Case _Registered","Case_Chargesheeted_%"]]
```

```python
cvt =df.groupby('State_UT')['Case_Chargesheeted_%'].sum()
cvt.head()
```

### Pie Chart Reprentation  State-Wise % of Case_Chargesheeted

```python

plt.figure(figsize=(8,8))
cvt.plot.pie(autopct="%1.2f%%")
plt.title("State-Wise % of Case_Chargesheeted")

plt.show()
```

### Viewing State with  Maximum % Of Case_Convicted 

```python
df[df['Case_Convicted_%']==(df["Case_Convicted_%"].max())][['State_UT',"Case_Convicted_%"]]
```

### Viewing State with  Minimum % Of Case_Convicted 

```python
df[df['Case_Convicted_%']==(df["Case_Convicted_%"].min())][['State_UT',"Case _Registered","Case_Convicted_%"]]
df.head()
```

```python
cvt1 =df.groupby('State_UT')['Case_Convicted_%'].sum()
cvt1.head()
```

### Pie Chart Reprentation  State-Wise % of Case_Case_Convicted

```python
plt.figure(figsize=(8,8))
cvt1.plot.pie(autopct="%1.2f%%")
plt.title("State-Wise % of Case_Convicted")

plt.show()
```

```python
all= df[["Case _Registered","Cases_Chargesheeted","No of Police_Chargesheeted","Cases_Final Report Submitted","No of Police_Discharged",
           "Cases_ Stayed by Courts",
           "No of Police Personel_Convicted"]].sum()
all
```

### Arranging in  Decreasing Order W.R.T  Cases_Chargesheeted

```python
data=df.sort_values(by="Cases_Chargesheeted", ascending=False)
data.head()
```

```python
x= data["Case _Registered"]
y= data["Cases_Chargesheeted"]
x1= data["Case _Registered"]
y1= data["No of Police_Chargesheeted"]

plt.xlabel("Case _Registered")
plt.ylabel("Count ")
plt.plot(x,y, color="red", label="Cases_Chargesheeted")
plt.plot(x,y1, color="blue", label="No of Police_Chargesheeted")
plt.legend()
```
